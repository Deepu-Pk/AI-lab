```
#Importing all essential library
import math
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def logLikelihoodFunction(mean, variance,data):#fuction to compute loglikelihood
    n= len(data)
    constant = -0.5*n*math.log(2*math.pi) - n*math.log(variance)
    remaining = 0
    for i in range(n):
        remaining += math.pow(data[i]-mean,2)/(2*math.pow(variance,2))*-1
    return constant+remaining

mean = 0 #Mean of uniform distribution
variance = 1 #Variance of uniform distribution
data = np.random.normal(mean, variance, 50) #Collectind data 
figure, ax = plt.subplots(1, figsize=(12,7))
count, bins, ignored = plt.hist(data,30, density=True)
ax.plot(bins,norm.pdf(bins,mean,variance))
plt.show()
np.mean(data)
np.std(data)

#Print likelihood by using diffent variance and mean of uniform distribution
print(logLikelihoodFunction(-4,2,data))
print(logLikelihoodFunction(0,1,data))
print(logLikelihoodFunction(2,1,data))
print(logLikelihoodFunction(-2,1,data))


figure, ax = plt.subplots(1, figsize=(10,8))
count, bins, ignored = ax.hist(data,30, density=True, color='red')

#we already proved that Maximum-Likelihood Estimate of mu is the mean of
# the measurements and for sigma = standard deviation.
estimated_mu = np.mean(data)
estimated_sigma= np.std(data)

# plotting the PDF of Estimated Distribution.
ax.plot(bins, norm.pdf(bins,estimated_mu,estimated_sigma),'--', lw=2.5,c='b',
label='mean={}, std={}'.format(round(np.mean(data), 3),
round(np.std(data),3)))

# plotting the PDF of True Distribution
ax.plot(bins, norm.pdf(bins,0,1),'-.', lw=2,c='g',
label='mean={}, std={}'.format(0,1))

ax.plot(bins, norm.pdf(bins,-1,1), lw=2,c='green',
label='mean={}, std={}'.format(0,1))
plt.legend()
plt.show()
```
